### Remove Unwanted Observations
Since GWA codes are aggreagtions of the regional MWA codes, i'll be using the GWA codes to perform this analysis.


```python
# Get all GWA codes
print('Before:', len(crypto_df))
gwa_codes = [Code for Code in crypto_df.Code.unique() if 'GWA_' in Code]
```

```python
#Kepp only GWA observations
crypto_df= crypto_df[crypto_df.Code.isin(gwa_codes)]
print(crypto_df)
```

```python
print('After:', len(crypto_df))
```

Next is to pivot the dataset and rearrange the table making the pivot the V-WAP column.

```python
pivoted_crypto_df = crypto_df.pivot(index ='Date', columns ='Code', values='V-WAP')
pivoted_crypto_df.tail()
```

```python
#Calculate returns over 7days
returns_7 = pivoted_crypto_df/pivoted_crypto_df.shift(7)-1.0
returns_7.tail()
```

```python
#Calculate returns for 7, 14,21,28 days

# Write a For loop and store the in a dictionary

#Dateoffsets are a standard kind of date increment used for a date range in Pandas

return_dict= {}
for offset in [7,14,21,28]:
    return_dict['return_{}'.format(offset)] = pivoted_crypto_df/pivoted_crypto_df.shift(offset)-1.0
```

By unpivoting, or melting the data, we can later create an analytical base table (ABT) where each row contains all of the relevant information for a particular coin on a particular date.


```python
#Melting the shifted data
melted_7 = returns_7.reset_index().melt(id_vars=['Date'], value_name='Returns_7')

# Melted dataframe examples
melted_7.tail()
```

```python
# Melt all the delta dataframes and store in list
melted_dfs = []
for key, return_df in return_dict.items():
    melted_dfs.append( return_df.reset_index().melt(id_vars=['Date'], value_name=key) )
```

create another melted dataframe that contains the forward-looking 7-day returns. This will be our "target variable" for evaluating our trading strategy.

```python
# Calculate 7-day returns after the date
forward_return_df = pivoted_crypto_df.shift(-7) / pivoted_crypto_df - 1.0
 
# Melt the return dataset and append to list
melted_dfs.append( forward_return_df.reset_index().melt(id_vars=['Date'], value_name='forward_return_7') )
```

join the melted dataframes into a single analytical base table. 

```python
# Merge two dataframes
pd.merge(melted_dfs[0], melted_dfs[1], on=['Date', 'Code']).tail()
```

Melting dataframes together and adding new features using the python built in function, reduce.


```python
from functools import reduce
```

```python
# Grab features from original dataset
base_df = crypto_df[['Date', 'Code', 'Volume', 'V-WAP']]
 
# Create a list with all the feature dataframes
feature_dfs = [base_df] + melted_dfs

# Reduce-merge features into analytical base table
abt = reduce(lambda left,right: pd.merge(left,right,on=['Date', 'Code']), feature_dfs)
 
# Display examples from the ABT
abt.tail(10)
```

Technically, with this ABT, we can already answer our original objective. For example, if we wanted to pick the coin that had the biggest momentum on January 3rd, 2018, we could simply display the rows for that date and look at the 7, 14, 21, and 28-day prior returns

```python
# Data from Jan 3rd, 2018
abt[abt.Date == '2018-01-03']
```

Checking for the crypto with the highest momentum.


```python
# Create 'month' feature
abt['month'] = abt.Date.apply(lambda x: x[:7])
 
# Group by 'Code' and 'month' and keep first date
gb_df = abt.groupby(['Code', 'month']).first().reset_index()
 
# Display examples
gb_df.tail()
```
